```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('country_wise_latest.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
sns.scatterplot(x='Confirmed',y='Recovered',data=df,hue='WHO Region')
plt.legend(bbox_to_anchor=(1.05,1));
```

```python
sns.scatterplot(data=df,x='Confirmed',y='Deaths',hue='WHO Region')
plt.legend(bbox_to_anchor=(1.05,1));
```

```python
sns.scatterplot(data=df,y='New recovered',x='New cases',hue='WHO Region')
plt.legend(bbox_to_anchor=(1.05,1));
```

```python
sns.scatterplot(data=df,y='New deaths',x='New cases',hue='WHO Region')
plt.legend(bbox_to_anchor=(1.05,1));
```
